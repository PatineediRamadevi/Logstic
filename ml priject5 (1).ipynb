```python

d2=pd.get_dummies(d1)
d2
```

```python
y=d2["Churn"]
x=d2.drop("Churn",axis=1)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
reg=LogisticRegression() # try the same without dropping 
```

```python
reg.fit(x_train,y_train)
```

```python
ypred=reg.predict(x_test)
ypred
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,ypred)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,ypred) 
```

```python
results=pd.DataFrame(columns=["original","predicted"])
```

```python

from sklearn import metrics
ypred = reg.predict_proba(x_test)[::,1]
fpr, tpr, _ = metrics.roc_curve(y_test,ypred)

#create ROC curve
plt.plot(fpr,tpr)
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```
